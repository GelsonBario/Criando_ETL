# Criando uma ETL com Python por Gelson Bario


```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
file_name = next(iter(uploaded))
df_rh = pd.read_csv(file_name)
print(f"Arquivo '{file_name}' carregado e extraido com sucesso")
print(df_rh.head())

from datetime import datetime

# Definindo a data de refer√™ncia (pode ser a data de hoje no Colab)
DATA_DE_HOJE = datetime(2025, 12, 1) # Mantenho a data fixa para consist√™ncia do exemplo

# ... (Seu c√≥digo de Limpeza e C√°lculo de Tempo de Servi√ßo) ...
df_rh['data_admissao'] = pd.to_datetime(df_rh['data_admissao'])
df_rh['tempo_servico_anos'] = ((DATA_DE_HOJE - df_rh['data_admissao']).dt.days / 365.25).round(2)

# ... (Sua fun√ß√£o de Classifica√ß√£o de Risco) ...
def classificar_risco_turnover(row):
    # L√≥gica de Neg√≥cio (risco alto se novo e com nota baixa)
    if row['tempo_servico_anos'] < 1.5 and row['nota_avaliacao_anual'] < 3.5:
        return 'ALTO RISCO üî•'
    # ... (outras regras) ...
    elif row['tempo_servico_anos'] > 5:
        return 'EST√ÅVEL (Veterano)'
    else:
        return 'NORMAL'

df_rh['risco_turnover'] = df_rh.apply(classificar_risco_turnover, axis=1)

print("ETAPA T: Transforma√ß√£o conclu√≠da. Novas colunas adicionadas.")
print(df_rh[['nome', 'tempo_servico_anos', 'risco_turnover']].head().to_markdown(index=False))

# --- ETAPA 3: CARGA (L) ---

# Salva o DataFrame final na m√°quina virtual do Colab
df_rh_final = df_rh[['id_funcionario', 'departamento', 'tempo_servico_anos', 'risco_turnover']]
df_rh_final.to_csv('relatorio_rh_transformado.csv', index=False)

# Faz o download do arquivo transformado para o seu computador local
files.download('relatorio_rh_transformado.csv')

print("\nETAPA L: O arquivo 'relatorio_rh_transformado.csv' foi salvo na m√°quina virtual e o download foi iniciado para o seu computador.")
```
